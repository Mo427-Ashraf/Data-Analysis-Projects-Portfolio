### Task2: Exploratory Data Analysis(Stock Prices Dataset)
##### "Best Practice"

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("Stocks shape:", df.shape)
print("\nStocks dtypes:\n", df.dtypes)
print("\nStocks head:\n", df.head())

# ==== Summary stats ====
print("\nDescribe (numeric):\n", df.describe())
print("\nMedian (numeric):\n", df.median(numeric_only=True))
print("\nMode (row 0):\n", df.mode().iloc[0])

# ==== Columns ====
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
print("\nNumeric columns:", num_cols)
print("Categorical/other columns:", cat_cols)







```

```python

# ==== Histograms for key price columns if present ====
for col in ["open", "high", "low", "close", "volume"]:
    if col in df.columns:
        plt.figure()
        df[col].hist(bins=30)
        plt.title(f"Stocks  Histogram of {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.show()
```

```python
#Boxplots for price columns
for col in ["open", "high", "low", "close"]:
    if col in df.columns:
        plt.figure()
        plt.boxplot(df[col].dropna(), vert=True)
        plt.title(f"Stocks  Boxplot of {col}")
        plt.ylabel(col)
        plt.show()
```

```python
#Scatter relationships
pairs = [
    ("open", "close"),
    ("high", "low"),
    ("close", "volume")
]
for x, y in pairs:
    if x in df.columns and y in df.columns:
        plt.figure()
        plt.scatter(df[x], df[y])
        plt.title(f"Stocks  Scatter: {x} vs {y}")
        plt.xlabel(x); plt.ylabel(y)
        plt.show()
```

```python
#visualize over time for a single symbol:
symbol_col = "symbol"
if symbol_col in df.columns and "close" in df.columns and "date" in df.columns:
    one_symbol = df[symbol_col].iloc[0]
    subset = df[df[symbol_col] == one_symbol].sort_values("date")
    plt.figure()
    plt.plot(subset["date"], subset["close"])
    plt.title(f"Stocks  Close over time ({one_symbol})")
    plt.xlabel("date"); plt.ylabel("close")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```

```python
# ==== Correlation heatmap (numeric only) ====
corr = df[num_cols].corr(numeric_only=True)
plt.figure()
plt.imshow(corr, interpolation="nearest")
plt.title("Stocks  Correlation Heatmap")
plt.colorbar()
plt.xticks(range(len(num_cols)), num_cols, rotation=45, ha="right")
plt.yticks(range(len(num_cols)), num_cols)
plt.tight_layout()
plt.show()
```

### Task3: Basic Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Bar Plot: Total Trading Volume per Symbol
plt.figure(figsize=(10,6))
df.groupby("symbol")["volume"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Top 10 Stocks by Total Trading Volume")
plt.xlabel("Stock Symbol")
plt.ylabel("Total Volume")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("barplot_volume.png")
plt.show()


```

```python
#2.Line Chart: Closing Price Over Time
one_stock = "F"
subset = df[df["symbol"] == one_stock].sort_values("date")

plt.figure(figsize=(12,6))
plt.plot(subset["date"], subset["close"], label=f"{one_stock} Closing Price")
plt.title(f"{one_stock} Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("linechart_close.png")
plt.show()
```

```python
#2.Line Chart: Closing Price Over Time
one_stock = "BAC"
subset = df[df["symbol"] == one_stock].sort_values("date")

plt.figure(figsize=(12,6))
plt.plot(subset["date"], subset["close"], label=f"{one_stock} Closing Price")
plt.title(f"{one_stock} Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("linechart_close.png")
plt.show()
```

```python

#3. Scatter Plot: Open vs Close Prices
plt.figure(figsize=(8,6))
sns.scatterplot(x="open", y="close", data=df, alpha=0.5)
plt.title("Open vs Close Prices")
plt.xlabel("Open Price")
plt.ylabel("Close Price")
plt.tight_layout()
plt.savefig("scatter_open_vs_close.png")
plt.show()
```
